# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Historgram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10,6))
plt.boxplot(df['Age'].dropna())
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

# Drop rows with missing values in either 'Age' or 'WorkWeekHrs'.
df = df.dropna(subset=['Age', 'WorkWeekHrs'])

plt.figure(figsize=(10,6))
plt.scatter(df['Age'], df['WorkWeekHrs'], edgecolor='w')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

# Drop rows with missing values in either 'Age', 'CodeRevHrs' or 'WorkWeekHrs'.
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

plt.figure(figsize=(10,6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, edgecolor='w')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble Size = Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# print all the tables names in the database
QUERY = """
SELECT * FROM
DatabaseDesireNextYear
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count FROM
DatabaseDesireNextYear GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8,8))
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next year')
plt.show()
```

```python
# print all the tables names in the database
QUERY = """
SELECT * FROM
LanguageDesireNextYear
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
WITH RankedLanguages AS (
    SELECT LanguageDesireNextYear, 
           RANK() OVER (ORDER BY COUNT(*) DESC) AS Rank
    FROM LanguageDesireNextYear
    GROUP BY LanguageDesireNextYear
)
SELECT Rank
FROM RankedLanguages
WHERE LanguageDesireNextYear = 'Python';
"""

df = pd.read_sql(QUERY, conn)

python_rank = df['Rank'].iloc[0] if not df.empty else 'Not found'
print(f"The rank of Python is: {python_rank}")
```

```python
# print all the tables names in the database
QUERY = """
SELECT * FROM
LanguageWorkedWith
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT COUNT(*) AS NumberOfRespondents
FROM LanguageWorkedWith
WHERE LanguageWorkedWith = 'SQL';
"""
df = pd.read_sql(QUERY, conn)

number_of_respondents = df['NumberOfRespondents'].iloc[0] if not df.empty else 'No data'
print(f"Number of respondents working with SQL: {number_of_respondents}")
```

```python
QUERY = """
SELECT COUNT(DISTINCT Respondent) AS NumberOfRespondents
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%mysql%';
"""
df = pd.read_sql(QUERY, conn)

number_of_respondents = df['NumberOfRespondents'].iloc[0] if not df.empty else 'No data'
print(f"Number of respondents working with MySQL: {number_of_respondents}")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs 
FROM master WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY,conn)

median_values = df[['WorkWeekHrs', 'CodeRevHrs']].median()

plt.figure(figsize=(8,6))
plt.bar(median_values.index, median_values['WorkWeekHrs'], label='Median WorkWeekHrs', color='b')
plt.bar(median_values.index, median_values['CodeRevHrs'], bottom=median_values['WorkWeekHrs'], label='Median CodeRevHrs', color='r')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.xlabel('Metrics')
plt.ylabel('Median Hours')
plt.legend()
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
WITH OrderedComp AS (
    SELECT Age,
           ConvertedComp,
           ROW_NUMBER() OVER (PARTITION BY Age ORDER BY ConvertedComp) AS RowAsc,
           ROW_NUMBER() OVER (PARTITION BY Age ORDER BY ConvertedComp DESC) AS RowDesc,
           COUNT(*) OVER (PARTITION BY Age) AS TotalCount
    FROM master
    WHERE Age BETWEEN 45 AND 60
),
MedianComp AS (
    SELECT Age,
           AVG(ConvertedComp) AS MedianConvertedComp
    FROM OrderedComp
    WHERE RowAsc = RowDesc
       OR RowAsc + 1 = RowDesc
    GROUP BY Age
)
SELECT Age, MedianConvertedComp
FROM MedianComp
ORDER BY Age;
"""

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10,6))
plt.plot(df['Age'], df['MedianConvertedComp'], marker='o', color='b')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(range(45,61)) #Set x-ticks for each age in the range
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch FROM
master
"""
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Count 
FROM master
GROUP BY MainBranch
ORDER BY Count DESC;
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10,6))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')
plt.title('Count of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
